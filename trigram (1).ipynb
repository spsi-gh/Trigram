```python
import torch.nn.functional as F

xenc = F.one_hot(xs, num_classes=27).float()
xenc = torch.flatten(xenc, 1)
xenc.shape
```

```python
nlls = torch.zeros(4)
for i in range(4):
  x1 = xs[i][0].item()
  x2 = xs[i][1].item()
  y = ys[i].item()
  print(f"The Actual Trigram {i+1}: {itos[x1]}{itos[x2]}{itos[y]}")
  print(f"Input to Neural Network: {x1}{x2}")
  print(f"Desired Output: {y}")
  print(f"Probabilities Assigned: {probs[i]}")
  p = probs[i, y]
  print(f"Probability Assigned to the Actual Output: {p}")
  logprob = torch.log(p)
  print(f"Loglikelihood: {logprob.item()}")
  nll = -logprob
  print(f"Negative Loglikelihood: {nll.item()}\n")
  nlls[i] = nll

print(f"Average nll: {nlls.mean()}")
```

```python
g = torch.Generator().manual_seed(2147483647)
W = torch.randn((54, 27), generator=g, requires_grad= True)
```

```python
for k in range(10):
  xenc = F.one_hot(xs, num_classes= 27).float()
  xenc = torch.flatten(xenc, 1)
  logits = xenc @ W
  #softmax
  counts = logits.exp()
  prob = counts/counts.sum(dim= 1, keepdims= True)
  loss = -prob[torch.arange(196113), ys].log().mean() + 0.01*(W**2).mean()

  print(loss.item())

  W.grad = None

  loss.backward()

  W.data += -10.0*W.grad
```

```python
for i in range(10):
  out = []
  ix1, ix2 = 0, 0

  while True:
    xenc = F.one_hot(torch.tensor([ix1, ix2]), num_classes=27).float()
    xenc = xenc.view(1, -1) # Reshape from [2, 27] to [1, 54]
    logits = xenc @ W
    counts = logits.exp()
    prob = counts / counts.sum(dim=1, keepdims=True)

    ix3 = torch.multinomial(prob, num_samples=1, replacement=True).item()

    ix1 = ix2
    ix2 = ix3

    out.append(itos[ix3])

    if ix3 == 0:
      break

  print(''.join(out))
```
